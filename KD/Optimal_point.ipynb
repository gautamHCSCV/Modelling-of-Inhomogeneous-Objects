````python
df = pd.read_csv(path+'100_max_depth_optimal_force_ecoflex_181221_derivative_data.csv')
df.head()
````

````python
mini, location = 25, -1
for i in range(1,201):
    arr = np.array(pd.read_csv(path+f'{i}_max_depth_optimal_force_ecoflex_181221_derivative_data.csv'))
    err = []
    for j in range(1,201):
        arr2 = np.array(pd.read_csv(path+f'{j}_max_depth_optimal_force_ecoflex_181221_derivative_data.csv'))
        m = min(len(arr),len(arr2))
        err.append(mean_squared_error(arr[:m], arr2[:m])**0.5)
    avg = np.mean(np.array(err))
    if avg<mini:
        mini = avg
        location = i
    with open('optimal_point.txt', 'a') as f:
        f.write(f'Point: {i}, Avg Error: {avg}, std: {np.std(np.array(err))}\n')
print(mini, location)
````
